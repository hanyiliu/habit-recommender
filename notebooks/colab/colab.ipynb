Setup SSH key to copy from private repo:

```python
import os
import subprocess

# 1) Create .ssh directory
os.makedirs("/root/.ssh", exist_ok=True)

# 2) Paste your private key between the triple quotes (exactly, no extra spaces)
private_key = """-----BEGIN OPENSSH PRIVATE KEY-----
[haha]
-----END OPENSSH PRIVATE KEY-----
"""

# 3) Write the key file
with open("/root/.ssh/id_ed25519", "w") as f:
    f.write(private_key)

# 4) Restrict permissions
subprocess.run(["chmod", "600", "/root/.ssh/id_ed25519"])

# 5) Add GitHub to known_hosts so SSH won't ask to confirm
known_hosts = "/root/.ssh/known_hosts"
subprocess.run(
    ["ssh-keyscan", "github.com"],
    stdout=open(known_hosts, "w"),
    check=True,
)
subprocess.run(["chmod", "644", known_hosts])
```
